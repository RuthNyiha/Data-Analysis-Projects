```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/product_sales.csv')
df.head()
```

# Data validation

##  Handling missing values by imputing with mean

```python
# Calculate the mean of the revenue column
revenue_mean = df['revenue'].mean()

# Fill null values in the revenue column with the mean
df['revenue'].fillna(revenue_mean, inplace=True)
```

# Exploratory Analysis

```python
# Ensure data types are correct
df['week'] = pd.to_numeric(df['week'], errors='coerce')
```

```python
# Create subplots with 3 rows and 2 columns
fig, axes = plt.subplots(3, 2, figsize=(16, 18))

# First row, first column: Distribution of Sales Methods
sns.countplot(x='sales_method', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Sales Methods')

# First row, second column: Distribution of Revenue
sns.histplot(df['revenue'], bins=20, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Revenue')

# Second row, first column: Distribution of Number of Sold Items
sns.histplot(df['nb_sold'], bins=20, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Number of Sold Items')

# Second row, second column: Scatterplot of Revenue vs. Site Visits by Sales Method
sns.scatterplot(x='nb_site_visits', y='revenue', hue='sales_method', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Scatterplot of Revenue vs. Site Visits by Sales Method')

# Third row, first column: Plotting sales method vs. number of items sold (nb_sold)
sns.barplot(x='sales_method', y='nb_sold', data=df, ax=axes[2, 0])
axes[2, 0].set_title('Sales Method vs. Number of Items Sold')
axes[2, 0].set_xlabel('Sales Method')
axes[2, 0].set_ylabel('Number of Items Sold')

# Third row, second column: Plotting sales method vs. revenue
sns.barplot(x='sales_method', y='revenue', data=df, ax=axes[2, 1])
axes[2, 1].set_title('Sales Method vs. Revenue')
axes[2, 1].set_xlabel('Sales Method')
axes[2, 1].set_ylabel('Revenue')

plt.tight_layout()  # Adjust layout to prevent overlapping of subplots
plt.show()
```
